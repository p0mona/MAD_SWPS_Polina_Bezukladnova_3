```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
apple = pd.read_csv("/content/apple_quality.csv")
```

```python
data = apple.head(15)
df = pd.DataFrame(data)
print(df)
```

```python
plt.hist(df['Juiciness'],bins = 20, color='skyblue')
plt.title('Distribution of Juiciness')
plt.xlabel('Juiciness')
plt.show()
```

```python
plt.plot(df['Crunchiness'], df['Sweetness'], 'r^')
plt.title('Crunchiness vs Sweetness')
plt.xlabel('Crunchiness')
plt.ylabel('Sweetness')
plt.show()
```

```python
quality_numeric = df['Quality'].apply(lambda x: 1 if x == 'good' else 0)
plt.plot(df['Acidity'], quality_numeric, marker='o')
plt.title('Quality vs Acidity')
plt.xlabel('Acidity')
plt.ylabel('Quality (1=good, 0=bad)')
plt.show()
```

```python
count = df['Quality'].value_counts()
plt.pie(count, labels=count.index, autopct='%1.1f%%', colors=['skyblue', 'pink'], startangle=90)
plt.title('Quality Distribution')
plt.show()
```
